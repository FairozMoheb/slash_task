```python
import pandas as  pd
import matplotlib.pyplot as plt
```

```python
report = pd.read_csv("amazon_sales_report.csv", encoding="latin-1")
report.head()
```

```python
report.dtypes
```

```python
#summary statistics for numerical variables
report.describe()
```

```python
#summary statistics for numerical variables
report.describe(include=['object'])
```

```python
report['Status'].value_counts().plot(kind='bar')
plt.xlabel('Status')
plt.ylabel('Frequency')
plt.title('Category Distribution')
plt.show()
```

```python

data = report['Status'].value_counts(normalize=True)*100

plt.figure(figsize=(8, 8))
plt.pie(data.values.tolist(), labels=data.index.tolist())
plt.title('TODO')
plt.show()
```
